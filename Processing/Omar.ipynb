```python
import pandas as pd
import numpy as np
import datetime
datetime.datetime.strptime
```

```python
#import CSV
reviews_csv = "AirBNB_Reviews.csv"

reviews_df = pd.read_csv(reviews_csv, encoding='latin-1')
```

```python
reviews_df.head()
```

```python
reviews_df.shape
```

```python
#check for null values
reviews_df.isnull().any()
```

```python
#Dropping rows with no reviews
clean_reviews_df= reviews_df[reviews_df.comments.notnull()]
```

```python
clean_reviews_df.isnull().any()
```

```python
new_df = clean_reviews_df.reset_index()

```

```python
new_df.head()
```

```python
#remove punctuation from comments
import string

def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

new_df["new_column"] = new_df['comments'].apply(remove_punctuations)
```
